### Filtre du second ordre



$V^-=\cfrac{\cfrac{V_e}{R_1}+\cfrac{V_1}{R_2}+\cfrac{V_3}{R_3}+\cfrac{V_s}{R_4}}{\cfrac{1}{R_1}+\cfrac{1}{R_2}+\cfrac{1}{R_3}+\cfrac{1}{R_4}}=0$

$V_1=V_2$

$V_3=\cfrac{V_2}{\tau p}$

$V_s=\cfrac{V_3}{\tau p}$

$\cfrac{V_e}{R_1}+\cfrac{V_s(\tau p)^2}{R_2}+\cfrac{V_s\tau p}{R_3}+\cfrac{V_s}{R_4}=0$

$V_eR_2R_3R_4+V_s(R_1R_3R_4(\tau p)^2+R_1R_2R_4\tau p+R_1R_2R_3)=0$

$\cfrac{V_s}{V_e}=-\cfrac{R_2R_3R_4}{R_1R_3R_4(\tau p)^2+R_1R_2R_4\tau p+R_1R_2R_3}=\cfrac{-\cfrac{R_4}{R_1}}{1+\cfrac{R_4}{R_3}\tau p+\cfrac{R_4}{R_2}(\tau p)^2}$

$\cfrac{1}{1+2\xi\cfrac{p}{\omega_0}+\left(\cfrac{p}{\omega_0}\right)^2}$

```python
from sympy.solvers import solve
from sympy import Symbol
R42 = Symbol('\cfrac{R_4}{R_2}')
f = 12000
w = 2*pi.evalf()*f
R42 = solve(R42*tau**2*w**2-1, R42)[0]
R42
```

```python
R43 = Symbol('\cfrac{R_4}{R_3}')
xi = 1
R43 = solve(R43*tau*w/(2*xi)-1, R43)[0]
R43
```

```python
E12 = [10,12,15,18,22,27,33,47,56,68,82]
for R4 in E12:
    R2 = R4/R42
    R3 = R4/R43
    perf_R2 =  min([abs(R2-R)/R2 for R in E12]+[abs(R2-10*R)/R2 for R in E12]+[abs(R2-R/10)/R2 for R in E12])*100
    perf_R3 =  min([abs(R3-R)/R3 for R in E12]+[abs(R3-10*R)/R3 for R in E12]+[abs(R3-R/10)/R3 for R in E12])*100
    print 'R4:%i\tR2: %i\tR3: %i\t%i' % (R4*1000, R2*1000, R3*1000, perf_R2 + perf_R3)
```

## Synthèse du filtre anti-repliement à capacités commutées

### Dimensionnement

Butterworth est le plus sélectif sans ondulations, et il possède un retard constant.

```python
20*log10(1/2**8)
```

Il nous faut donc 50 dB d’atténuation à $F_e = 100kHz$, mais on utilise un ordre 4 ce qui nous permet de pouvoir autoriser une fréquence de coupure de 25kHz, afin de ne pas atténuer les fréquences proches de la bande passante du capteur.

```python
R431 = Symbol('R4/R3')
R431 = solve(R431*tau*w/1.848-1, R431)[0]
R431
```

```python
R421 = Symbol('R4/R2')
R421 = solve(R421*tau**2*w**2-1, R421)[0]
R421
```

```python
for R4 in E12:
    R2 = R4/R421
    R3 = R4/R431
    perf_R2 =  min([abs(R2-R)/R2 for R in E12]+[abs(R2-10*R)/R2 for R in E12]+[abs(R2-R/10)/R2 for R in E12])*100
    perf_R3 =  min([abs(R3-R)/R3 for R in E12]+[abs(R3-10*R)/R3 for R in E12]+[abs(R3-R/10)/R3 for R in E12])*100
    print 'R4:%i\tR2: %i\tR3: %i\t%i' % (R4*1000, R2*1000, R3*1000, perf_R2 + perf_R3)
```
