Si on n’utilise pas le générateur, on peut aussi faire un MOS en empilant les bonnes couches. Cependant, lorsque nous essayons de faire un transistor de taille minimale ($W=L=2\lambda$), le DRC nous informe que $W$ est trop petit; on doit donc passer à $W=2L$.

On passe ensuite au design d’un NMOS de taille $W=10\mu m$, dont voici une coupe dans la vue 3D:

```python
Image(filename='microwind/nmos3d.png')
```

On peut alors le simuler, et afficher sa caractéristique $I_d(V_{ds})$. On utilise pour cela le modèle BSIM4, plus récent que Level 1 et Level 3:

```python
Image(filename='microwind/nmoscarac.png')
```

Ce transistor peut délivrer un courant maximal de $6.4mA$ si l’on impose $V_d=2.5$ et $V_g=2.5$.

La tension de seuil est visible sur le graphe $I_d(V_g)$, et vaut, pour $V_b=0$, 0.45, ce qui correspond au VTH0 du modèle BSIM4.

```python
Image(filename='microwind/nmosseuil.png')
```

Le courant de fuite est visible sur le graphe $log(I_d(V_g))$ et vaut environ $70nA$ pour $V_b=0$. On le mesure en regardant l’intersection de la courbe d’$I_d$ pour $V_b=0$ avec $V_{gs}=0$.

```python
Image(filename='microwind/nmosfuite.png')
```

Concernant l’évolution des modèles, on remarque que le Level 1 nous empêche simplement de calculer le courant fuite, puisqu’il n’est pas assez précis. Si l’on regarde $I_d(V_d)$, la différence entre BSIM et les Level 1 et 3 est également flagrante dans la zone linéaire.

Si cette tension de coupure n’est pas à $\cfrac{V_{dd}}{2}$, c’est que la mobilité des électrons et des trous n’est pas la même. Pour compenser ceci, il suffit de modifier le rapport $\cfrac{W}{L}$:

```python
Image(filename='microwind/inv4x.png')
```

On vérifie bien que la tension de coupure est à $\cfrac{V_{dd}}{2}$ après cette compensation:

```python
Image(filename='microwind/inv4xrelatif.png')
```

On peut ensuite comparer notre schéma avec celui de la librairie, qui ressemble à ça:

```python
Image(filename='microwind/invlib.png')
```

Cependant, on remarque que le rapport $\cfrac{W}{L}$ de l’inverseur de la lib est de 3, ce qui ne permet pas d’avoir une tension de coupure exactement à $\cfrac{V_{dd}}{2}$ comme le notre:

```python
Image(filename='microwind/invrelatifcomp.png')
```

On cable ensuite trois de ces inverseurs en série:

```python
Image(filename='microwind/inv3x.png')
```

Ce système oscille alors à $6.25GHz$, ce qui correspond à peu près à $\cfrac{1}{6(t_r + t_f)}$

```python
Image(filename='microwind/inv3xfreq.png')
```

En passant à 7 inverseurs en série, on oscille à $2.74GHz$. On a donc logiquement $\cfrac{6.25}{2.74}\simeq\cfrac{7}{3}$, vu que c’est le temps de propagation dans toute la boucle qui joue sur sa fréquence d’oscillation.

```python
(7/3-625/274)/(7/3)*100  # Pourcentage d’erreur entre 6.25/2.75 et 7/3
```

```python
Image(filename='microwind/inv7xfreq.png')
```

On remarque que le signal ressemble plus à un carré qu’à un sinus, c’est simplement parce que le temps de propagation est beaucoup plus long, et que le système avait déjà le temps de passer de $0V$ à $2.5V$ en $160ps$. Le reste du temps, il y reste donc, d’où cette ressemblance avec un signal «écrêté».

# Design d’une porte de transmission

Une porte de transmission sert à faire un interrupteur ; son schéma n’est pas bien sorcier:

```python
Image(filename='microwind/trans-sch.png')
```

Voici l’équivalent sous microwind:

```python
Image(filename='microwind/trans.png')
```

On peut également s’en servir en guise de mémoire, puisque le signal de sortie est en haute impédance si l’interrupteur est éteint, et donc sa tension à ce moment là est la tension appliquée au dernier moment où l’interrupteur était passant:

```python
Image(filename='microwind/transtemp.png')
```

La différence entre une porte avec un NMOS et un PMOS et un seul transistor, c’est qu’un PMOS ou un NMOS ne peuvent imposer véritablement un niveau haut et un niveau bas en fonction de l’entrée.
